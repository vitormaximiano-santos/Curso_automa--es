```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.by import By

# carregar excel
tabela = pd.read_excel("Treinos/Code_trabalho/Descargas_janeiro_2026.xlsx")

# remover linhas totalmente vazias
tabela = tabela.dropna(how="all")

# valores fixos
email_fixo = "Operacaosuzano@jsl.com.br"
transportadora_fixo = "JSL"
tipo_fixo = "Pedágio"
valor_espelho_fixo = 0

# arquivos de log
log_completas = open("dts_completas.txt", "w", encoding="utf-8")
log_incompletas = open("dts_incompletas.txt", "w", encoding="utf-8")

# iniciar navegador
navegador = webdriver.Chrome()

for i in range(len(tabela)):
    linha = tabela.loc[i]

    # capturar campos
    dt = linha.get("DT")
    qtd_eixos = linha.get("EIXO")
    valor_rota = linha.get("VALOR")

    # verificar campos obrigatórios
    campos_faltando = []
    if pd.isna(dt): campos_faltando.append("DT")
    if pd.isna(qtd_eixos): campos_faltando.append("EIXO")
    if pd.isna(valor_rota): campos_faltando.append("VALOR")

    if campos_faltando:
        log_incompletas.write(f"Linha {i+2} ignorada - campos faltando: {', '.join(campos_faltando)}\n")
        continue

    # abrir formulário
    navegador.get("https://forms.office.com/r/cBcVBYa5mZ")
    time.sleep(2)

    try:
        # preencher transportadora
        navegador.find_element(By.XPATH, '//*[@id="question-list"]/div[1]/div[2]/div/span/input').send_keys(transportadora_fixo)

        # preencher email
        navegador.find_element(By.XPATH, '//*[@id="question-list"]/div[2]/div[2]/div/span/input').send_keys(email_fixo)

        # preencher DT
        navegador.find_element(By.XPATH, '//*[@id="question-list"]/div[3]/div[2]/div/span/input').send_keys(str(dt))

        # selecionar "Pedágio"
        navegador.find_element(By.XPATH, '//*[@id="question-list"]/div[4]/div[2]/div/div/div[2]/div/label/span[1]/input').click()
        time.sleep(1)

        # preencher quantidade de eixos
        navegador.find_element(By.XPATH, '//*[@id="question-list"]/div[5]/div[2]/div/span/input').send_keys(str(qtd_eixos))

        # preencher valor da rota
        navegador.find_element(By.XPATH, '//*[@id="question-list"]/div[6]/div[2]/div/span/input').send_keys(str(valor_rota))

        # preencher valor espelho
        navegador.find_element(By.XPATH, '//*[@id="question-list"]/div[7]/div[2]/div/span/input').send_keys(str(valor_espelho_fixo))

        # enviar
        navegador.find_element(By.XPATH, '//*[@id="form-main-content1"]/div/div/div[2]/div[3]/div/button').click()
        time.sleep(2)

        # registrar sucesso
        log_completas.write(f"Linha {i+2} enviada com sucesso - DT: {dt}, Eixos: {qtd_eixos}, Valor: {valor_rota}\n")

    except Exception as e:
        log_incompletas.write(f"Linha {i+2} falhou - erro Selenium: {e}\n")

# fechar navegador e arquivos
navegador.quit()
log_completas.close()
log_incompletas.close()
```
